задание 1

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
```

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
boston = load_boston()
boston.keys()
```

```python
data = boston["data"]
feature_names = boston["feature_names"]
```

```python
X = pd.DataFrame(data, columns=feature_names)
```

```python
target = boston["target"]
y = pd.DataFrame(target, columns=["price"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
R2 = r2_score(y_test, y_pred)
R2
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators =100, max_depth = 12, random_state=0)
```

```python
model.fit(X_train, y_train.values[:, 0])
```

```python
y_pred = model.predict(X_test)
```

```python
R2 = r2_score(y_test, y_pred)
R2
```

RandomForestRegressor работает точнее, так как в этом случае значение коэффициента детерминации R2 ближе к единице

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
probability = clf.predict_proba(X_test)
```

```python
probability
```

```python
y_pred_proba = probability[0:,1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
